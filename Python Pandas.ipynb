```python
#Lọc các cột theo pattern của tên cột
#Khi làm việc với dữ liệu lớn sẽ có những tình huống mà bạn bắt gặp các cột 
# thuộc về cùng một nhóm và chúng có chung một pattern. Chẳng hạn như về age sẽ có
# age_1, age_2, age_3,… Làm thế nào để bạn lọc ra được những biến này từ dữ liệu? 
# Chúng ta sẽ dùng hàm filter. Đây là hàm cực kỳ tiện ích khi lọc cột từ những bộ 
# dữ liệu lớn mà bạn sẽ thường xuyên sử dụng sau này.
df2 = pd.DataFrame({
    'name':['a', 'b', 'c', 'd', 'e'],
    'age_1':[1, 2, 3, 4, 5],
    'age_2':[3, 5, 7, 9 , 10],
    'age_3':[2, 5, 2, 5, 6]
})

df2.head()
```

```python
#Lựa chọn các cột bắt đầu là age thông qua hàm filter.
# Trong pandas thì axis=1 là làm việc với cột và axis=0 là làm việc với dòng. Giá trị của regex=^age 
#có nghĩa là lọc các cột có chuỗi ký tự là age đứng đầu.


df2.filter(regex="^age", axis = 1)
```

```python
# min, max, mean, median, sum
print(df['tax'].min(), df['tax'].max(), df['tax'].mean(), df['tax'].median(), df['tax'].sum())
```

```python
#HAM cut
#Hàm cut giúp ta phân chia giá trị của một trường liên tục vào những khoảng theo ngưỡng cắt.
#Kết quả trả ra là nhãn của từng khoảng mà chúng ta khai báo.
bins = [-999999, 250, 400, 999999]
labels = ['low', 'normal', 'high']
# low: -999999 <- 250
# normal: 250 <- 400
# high: 400 <- 999999
df['tax_labels'] = pd.cut(df['tax'], bins=bins, labels=labels)
df[df['tax_labels']=='high'].head()
```

```python
#Ham qcut
import numpy as np

labels = ['low', 'normal', 'high']
tax_labels = pd.qcut(df['tax'], q=3, labels=labels)
np.unique(tax_labels, return_counts = True)
```

```python
#Apply
#Ví dụ bên dưới ta muốn nhân đôi giá trị của tax thì có thể sử dụng hàm apply với lambda như sau:
df["tax"].apply(lambda x: 2*x).head
```

```python
#Ta cũng có thể áp dụng cho nhiều trường một lúc. Khi đó cần khai báo axis=1 để biết rằng ta đang áp dụng trên từng cột,
#nếu axis=0 thì sẽ áp dụng trên từng dòng.
df[['tax', 'medv']].apply(lambda x: 2*x, axis=1).head()
```

```python
#Map
#Bên dưới ta sẽ map các giá trị của trường df['tax_labels'] sang các giá trị tiếng Việt.
dict_tax = {
    "low":"thap",
    "normal":"tb",
    "high":"cao"
    
}

df['tax_labels'].map(dict_tax).head
```
